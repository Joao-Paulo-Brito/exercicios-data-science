## Segunda pergunta
Antes de responder a segunda pergunta, o que é correlação ?<br>
R= Correlação mede o grau de variação entre duas variáveis, indicando como uma variável pode ser associada a outra.<br><br>

. Positiva: quando uma variável X aumenta, a variável Y também tende a aumentar.<br>
. Negativa: quando uma variável X aumenta, a variável Y tende a diminuir.<br>
. Neutra: Quando não há relação aparente entre as variáveis X e Y.<br><br>

### O que é correlação de Pearson ?<br><br>

-> É uma técnica estatística para testar se há uma relação linear entre duas variáveis quantitativas (continuas)
<br>-> O coeficiente de correlação de Pearson pode ter um intervalo de valores de -1 a +1. Um valor de 0 indica que não á associação entre as variáveis, se o valor for maior que 0 indica uma associação positiva, e o valor menor que 0, indica uma associação negativa.<br><br>

E o comando usado para verificar o coeficiente de correlação de Pearson em um data frame é *.corr()*
<br><br>

## Quarta pergunta
Para responder a quarta pergunta, é necessário fazer uma filtragem para os aplicativos que são abertos mais vezes, mas com pouco tempo de uso, e para isso, é preciso fazer um agrupamento tendo os aplicativos como referência

```python
x = df.groupby('App').agg({'Usage (minutes)': 'sum', 'Times Opened': 'sum'})
```

```python
x
```

```python
x = x.reset_index()
```

```python
x = x.rename(columns={'Usage (minutes)': 'Usage (minutes)', 'Times Opened': 'Times Opened'})
```

```python
x
```

```python
import numpy as np
```

```python
# Gráfico de barras empilhadas
# Dados para o gráfico
y = np.arange(len(x['App']))  # Índices das categorias
width = 0.35  # Largura das barras

plt.figure(figsize=(10, 6))

plt.bar(y - width/2, x['Usage (minutes)'], width, label='Tempo de uso', color='blue')
plt.bar(y + width/2, x['Times Opened'], width, label='Qtd de vezes aberto', color='orange')

plt.xlabel('Aplicativos')
plt.ylabel('Valores')
plt.title('Aplicativos abertos mais vezes e com pouco tempo de uso')
plt.xticks(y, x['App'])
plt.legend()
plt.show()
```
